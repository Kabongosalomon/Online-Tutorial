# Study of common Non-Linear activation function 

# Sigmoid

The sigmoid activation function $\sigma(x)=1/(1+e^{-1})$ is an S-shaped function, transforming each value x into the range $[0, 1]$.

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.e**(-x))

t1 = np.arange(-50, 50, 0.1)
plt.plot(t1, sigmoid(t1))
plt.show()
```

# Hyperbolic tangent (tanh)

The hyperbolic tangent $tanh(x)=\frac{e^{2x}-1}{e^{2x}+1}$ activation function is an S-shaped function, transforming the values x into the range $[ − 1, 1]$.

```python
# def Hyperbolic_tang(x):
#     return np.tanh(x)

def Hyperbolic_tang(x):
    return (np.e**(2*x)-1)/(np.e**(2*x)+1)

t1 = np.arange(-50, 50, 0.1)
plt.plot(t1, Hyperbolic_tang(t1))
plt.ylabel("sigmoid")
plt.xlabel("x")
plt.title("Activation function : σ")
plt.show()
```

# Hard tanh

The hard-tanh activation function is an approximation of the tanh function
which is faster to compute and take derivatives of:

\begin{align*}
    hardtanh(x)&=
    \left\{
    \begin{array}{cr}
        -1&x<-1\\
        1&x>1\\
        x&\text{otherwise}
        \end{array}
    \right.
\end{align*}

```python
def hardtanh(x) : 
    return -1 if x<-1 else (1 if x>1 else x)
    
# hardtanh_ = lambda x: -1 if(x<-1) else if (x>1) else x

t1 = np.arange(-50, 50, 0.1)

plt.plot(t1, [hardtanh(t1[i]) for i in range(len(t1))])
# plt.plot(t1, hardtanh(t1).any())

plt.ylabel("hardtanh")
plt.xlabel("x")
plt.title("Activation function : hardtanh")
plt.show()
```

# Rectiﬁer (ReLU)

The Rectiﬁer activation function *(Glorot, Bordes, & Bengio, 2011)*, also known as the rectiﬁed linear unit is a very simple activation function that is easy to work with and was shown many times to produce excellent results. The ReLU unit clips each value $x < 0$ at $0$. Despite its simplicity, it performs well for many tasks, especially when combined with the dropout regularization technique 

\begin{align*}
    ReLU=max(0, x)&=
    \left\{
    \begin{array}{cl}
        0&x<0\\
        x&\text{otherwise}
        \end{array}
    \right.
\end{align*}

As a rule of thumb, **ReLU** units work better than **tanh**, and **tanh** works better than **sigmoid**.

```python
def ReLU(x) :
    return np.maximum(0,x)


t1 = np.arange(-50, 50, 0.1)

plt.plot(t1, ReLU(t1))
plt.ylabel("ReLU")
plt.xlabel("x")
plt.title("Activation function : ReLU")
plt.show()
```

The *technical advantages* of the **ReLU** over the **sigmoid** and **tanh** activation functions is that it does not involve *expensive-to-compute functions*, and more importantly that it does not **saturate**. The **sigmoid** and **tanh** activation are capped at 1, and the *gradients* at this region of the functions are *near zero*,
driving the entire gradient near zero. The *ReLU activation* does not have this problem, making it especially suitable for *networks with multiple layers*, which are susceptible to the **vanishing gradients problem** when trained with the *saturating units*.

## Leaky ReLU

But because of the horizontal line with 0 as the output, we can face serious issues sometimes. For instance, in the previous section we discussed a horizontal line, which is a constant with a derivative of 0 and therefore may become a bottleneck during training, as the weights will not easily get updated. To circumvent the problem, there was a new activation function proposed: Leaky ReLU, where the negative value outputs a slightly slanting line instead of a horizontal line, which helps in updating the weights through backpropagation effectively. 

Leaky ReLU is defined as

\begin{align*}
    LeakyReLU&=
    \left\{
    \begin{array}{cl}
        x&\text{if : }x>0\\
        \alpha x& \text{if : } x<0
        \end{array}
    \right.
\end{align*}

- with $\alpha$ a parameter that is defined as a small constant, say 0.005


```python
def ReLU(x, alp) :
    if x>0:
        return x
    else:
        return alp*x


t1 = np.arange(-50, 50, 0.1)

alp = 0.05

plt.plot(t1, [ReLU(t1[i], alp) for i in range(len(t1))])
plt.ylabel("LeakyReLU")
plt.xlabel("x")
plt.title("Activation function : Leaky ReLU")
plt.show()
```

# In addition

**The cube and tanh-cube activation functions are motivated by the desire to better capture interactions between different features. While these activation functions are reported to improve performance in certain situations, their general applicability is still to be determined**.

## Cube


suggested by *(Chen & Manning, 2014)*, the **Cube activation function** $g(x)=(x^3)$ look more effective than other non-linearities in a feed-forward network that was used to predict the actions in a greedy transition-based dependency parser

```python
def Cube(x) :
    return x**3


t1 = np.arange(-50, 50, 0.1)

plt.plot(t1, Cube(t1))
plt.ylabel("Cube")
plt.xlabel("x")
plt.title("Activation function : Cube")
plt.show()
```

## tanh cube

function $g(x) = tanh((x)^3 + x)$ was proposed by *(Pei et al., 2015)*, who found it to be more eﬀective than other non-linearities in a feed-forward network that was used as a component in a structured-prediction graph-based dependency parser.

```python
def tanh_cube(x) :
    return np.tanh(x**3 +x)


t1 = np.arange(-50, 50, 0.1)

plt.plot(t1, tanh_cube(t1))
plt.ylabel("tanh_cube")
plt.xlabel("x")
plt.title("Activation function : tanh_cube")
plt.show()
```

There are many more activation functions that can be used in a DNN and are available in Keras. A few other popular ones are tanh (hyperbolic tan activation), swish activation, elu (exponential linear unit), selu (scaled elu), and so on.
